```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt


font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 16}

matplotlib.rc('font', **font)
```

```python
num_points=1000
std=0.20
x=1*np.random.rand(num_points)-1.5
y=(x**3)-5*(x**2)-10*(x)+5+np.random.randn(num_points)*std
```

```python
# underfit calculate polynomial
uf=[]
of=[]
opt=[]
for models in range(5):
    indices=np.random.randint(0,100,size=100)
    z = np.polyfit(x[indices], y[indices], 1)
    uf.append(np.poly1d(z))

    # overfit calculate polynomia
    z = np.polyfit(x[indices], y[indices], 10)
    of.append(np.poly1d(z))

    # overfit calculate polynomia
    z = np.polyfit(x[indices], y[indices], 3)
    opt.append(np.poly1d(z))
```

```python
plt.figure(figsize=[10,10])
plt.scatter(x[indices],y[indices])
h1=plt.plot(np.sort(x),uf[0](np.sort(x)),'-r')
h2=plt.plot(np.sort(x),of[0](np.sort(x)),'-g')
#h3=plt.plot(np.sort(x),opt[0](np.sort(x)),'-k')
plt.plot(np.sort(x),uf[1](np.sort(x)),'--r')
plt.plot(np.sort(x),of[1](np.sort(x)),'--g')
#plt.plot(np.sort(x),opt[1](np.sort(x)),'--k')
plt.plot(np.sort(x),uf[2](np.sort(x)),'-.r')
plt.plot(np.sort(x),of[2](np.sort(x)),'-.g')
#plt.plot(np.sort(x),opt[2](np.sort(x)),'-.k')
plt.legend(['Data','Underfit','Overfit'])
plt.show()
plt.close()
```
