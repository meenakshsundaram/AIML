```python

import numpy as np 
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import metrics
```

```python
num_data_points = 20000
num_positive = np.random.randint(0,num_data_points)
num_negative = num_data_points - num_positive

truth = np.ones(num_data_points)
truth[num_positive:]=0
```

```python

fig=plt.figure()
ax= fig.add_axes([0,0,1,1])
for trial in range(10):
    pred = np.random.rand(num_data_points)
    metrics.RocCurveDisplay.from_predictions(truth,pred,ax=ax)
plt.show()
plt.close()
```
